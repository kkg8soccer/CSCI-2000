```python
#Katherine Koehler-Grassau 100559637
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from JSAnimation.IPython_display import display_animation

import urllib
from mpl_toolkits.basemap import Basemap
import csv
Category = []
Lat = []
Long = []
Cases = []

fig = plt.figure(figsize=(10,10))
ax = plt.subplot(1,1,1)

earth = Basemap(projection='mill')

earth.drawcoastlines(color='0.50', linewidth=0.25)
earth.fillcontinents(color='0.95')

with open('Source_data_for_CFR_vaccine_map_abridged.csv') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        Category.append(row['Category'])
        Lat.append((float)(row['Lat']))
        Long.append((float)(row['Long']))
        Cases.append((float)(row['Cases']))
       
list(map(float, Lat))
list(map(float, Long))
list(map(float, Cases))
k = 0
for i in range (k, len(Category)):
    if (Category[k] == 'Measles'):
        x, y = earth(Long[k], Lat[k])
       
        if (Cases[k] >= 0 and Cases[k] < 5000):
            earth.plot(x, y, 'ro',markersize=5,alpha=.5)
           
        elif (Cases[k] >= 5000 and Cases[k] < 140000):
            earth.plot(x, y, 'ro',markersize=((Cases[k])/1000),alpha=.5)
    k += 1
   
plt.title("Measles outbreaks (100556060)")
plt.show()

plt.figure(figsize=(10, 10))
earth = Basemap(projection='mill')
earth.drawcoastlines(color='0.50', linewidth=0.25)
earth.fillcontinents(color='0.95')
with open('Source_data_for_CFR_vaccine_map_abridged.csv') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        Category.append(row['Category'])
        Lat.append((float)(row['Lat']))
        Long.append((float)(row['Long']))
        Cases.append((float)(row['Cases']))
       
list(map(float, Lat))
list(map(float, Long))
list(map(float, Cases))
l = 0
for i in range (l, len(Category)):
    if (Category[l] == 'Rubella'):
        x, y = earth(Long[l], Lat[l])
       
        if(Cases[l] >= 0 and Cases[l] < 5000):
            earth.plot(x, y, 'mo',markersize=5,alpha=.5)
           
        elif(Cases[l] >= 5000 and Cases[l] < 140000):
            earth.plot(x, y, 'mo',markersize=((Cases[l])/1000),alpha=.5)
    l += 1     
   
plt.title("Rubella outbreaks (100556060)")
plt.show()

plt.figure(figsize=(10, 10))
earth = Basemap(projection='mill')
earth.drawcoastlines(color='0.50', linewidth=0.25)
earth.fillcontinents(color='0.95')
with open('Source_data_for_CFR_vaccine_map_abridged.csv') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        Category.append(row['Category'])
        Lat.append((float)(row['Lat']))
        Long.append((float)(row['Long']))
        Cases.append((float)(row['Cases']))
       
list(map(float, Lat))
list(map(float, Long))
list(map(float, Cases))
m = 0
for i in range (m, len(Category)):
    if (Category[m] == 'Mumps'):
        x, y = earth(Long[m], Lat[m])
       
        if (Cases[m] >= 0 and Cases[m] < 5000):
            earth.plot(x, y, 'go',markersize=5,alpha=.5)
           
        elif(Cases[m] >= 5000 and Cases[m] < 140000):
            earth.plot(x, y, 'go',markersize=((Cases[m])/1000),alpha=.5)
    m += 1     
   
plt.title("Mumps outbreaks (100556060)")
plt.show()

plt.figure(figsize=(10, 10))
earth = Basemap(projection='mill')
earth.drawcoastlines(color='0.50', linewidth=0.25)
earth.fillcontinents(color='0.95')
with open('Source_data_for_CFR_vaccine_map_abridged.csv') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        Category.append(row['Category'])
        Lat.append((float)(row['Lat']))
        Long.append((float)(row['Long']))
        Cases.append((float)(row['Cases']))
       
list(map(float, Lat))
list(map(float, Long))
list(map(float, Cases))
n = 0
for i in range (n, len(Category)):
    if (Category[n] == 'Polio'):
        x, y = earth(Long[n], Lat[n])
        if (Cases[n] >= 0 and Cases[n] < 5000):
            earth.plot(x, y, 'bo',markersize=5,alpha=.5)
           
        elif(Cases[n] >= 5000 and Cases[n] < 140000):
            earth.plot(x, y, 'bo',markersize=((Cases[n])/1000),alpha=.5)
    n += 1     
   
plt.title("Polio outbreaks (100556060)")
plt.show()
```
