```python
#Katherine Koehler-Grassau 100559637
from struct import *
import math as math
import matplotlib.pyplot as plt
import numpy as np

with open('halo.dat', 'rb') as file:
    N_halo = file.readline()
    T_halo = file.readline()
    data_halo = file.readline()
    
    struct.unpack("i", N_halo[:20])
    struct.unpack("i", T_halo[:-4])
    struct.unpack("i" * ((len(data_halo) -24) // 4), data_halo[20:-4])

Vx = []
Vy = []
Vz = []
v_halo = []
             
for i in range(len(data_halo)): # This part goes through the whole length of the array x
    v_halo.append(math.sqrt(Vx[i]**2 + Vy[i]**2 + Vz[i]**2))
    
    
    #!/usr/bin/env python
%matplotlib inline
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt

#Speed = np.loadtxt("v_halo")

mu, sigma = 100, 15
x = mu + sigma*np.random.randn(10000)

# the histogram of the data
n, bins, patches = plt.hist(v_halo, 50, normed=1, facecolor='green', alpha=0.75)

# add a 'best fit' line
y = mlab.normpdf( bins, mu, sigma)
l = plt.plot(bins, y, 'r--', linewidth=1)


plt.xlabel('Bodies')
plt.ylabel('Speeds')
plt.title("Velocity histogram (halo) 100559637")
plt.axis([40, 160, 0, 0.03])
plt.grid(True)

plt.show()

    
```
