```python
#Katherine Koehler-Grassau 100559637
from struct import *
import math as math
import matplotlib.pyplot as plt
import numpy as np # This is the necessary libraries for this question

with open('halo.dat', 'rb') as file:
    N_halo = file.readline() # This one reads the first line and puts it into the file N_halo
    T_halo = file.readline() # This one reads the second line and puts it into the file T_halo
    data_halo = file.read() #Reads the rest of the file into data_halo 
    
    #This one unpacks each of the files
    struct.unpack("i", N_halo[:20])
    struct.unpack("i", T_halo[:-4])
    struct.unpack("i" * ((len(data_halo) -24) // 4), data_halo[20:-4])

    # This creates the variables for the vectors 
    Vx = []
    Vy = [] 
    Vz = []
    v_halo = []
    
    # This one puts the data from each of the files into the vectors of vx, vy and vz
       i = 0
    for j in unpacked_data:
        if(i == 401661):
            break
        elif(i%3 == 0):
            vx.append(j)
            i = i + 1
        elif(i%3 == 1):
            vy.append(j)
            i = i + 1
        elif(i%3 == 2):
            vz.append(j)
            i = i + 1
   
    #This part calculates the speed using the data from vx, vy and vz and calcuates it using the given formula
    #sqare root of the components squared added together
    k = 0    
    while k < (i//3):
        for i in range(len(data_halo)): # this part goes thru the previous vectors
    v_halo.append(math.sqrt(Vx[i]**2 + Vy[i]**2 + Vz[i]**2))
             
    

#Speed = np.loadtxt("v_halo")

mu, sigma = 100, 15
x = mu + sigma*np.random.randn(10000)

# the histogram of the data, this plots the histogram with the data from v_halo and bins of 50
n, bins, patches = plt.hist(v_halo, 50, normed=1, facecolor='green', alpha=0.75)

# add a 'best fit' line
y = mlab.normpdf( bins, mu, sigma)
l = plt.plot(bins, y, 'r--', linewidth=1)

# This part describes the labelling for the graph that the user can change. Also is the length and width of the axes of the histogram.
plt.xlabel('Bodies')
plt.ylabel('Speeds')
plt.title("Velocity histogram (halo) 100559637")
plt.axis([40, 160, 0, 0.03])
plt.grid(True)

#This outputs the code into the plot.
plt.show()

    
```
