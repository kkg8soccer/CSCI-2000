```python
#Katherine Koehler-Grassau 100559637
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from JSAnimation.IPython_display import display_animation # libraries needed for this question

import urllib
from mpl_toolkits.basemap import Basemap
import csv #importing necessary files
# This one defines the vector variables
Cat = [] #Category
L1 = [] #latitude
L2 = [] #longitude
Cases = []

# Using the data we learned in class with the earthquake data I came up with a code.
#This plots the outbreak map for the measles
#This part determines the colours and line width of the plot
fig = plt.figure(figsize=(10,10))
ax = plt.subplot(1,1,1)

earth = Basemap(projection='mill')

#this allows the different colours for the continents and the coastlines
earth.drawcoastlines(color='0.50', linewidth=0.25)
earth.fillcontinents(color='0.95')

# This opens the necessary file with the data needed.
# it puts the data into vectors that are needed to plot the data.
with open('Source_data_for_CFR_vaccine_map_abridged.csv') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        Cat.append(row['Category'])
        L1.append((float)(row['Latitude']))
        L2.append((float)(row['Longitude']))
        Cases.append((float)(row['Cases']))

#This part goes through all the values and the vectors and runs through the for loop and elif if statements.
# Then based on the values inputted into the loop the proper colours and dots will be outputted.
list(map(float, L1))
list(map(float, L2))
list(map(float, Cases))
k = 0
for i in range (k, len(Cat)):
    if (Cat[k] == 'Measles'):
        x, y = earth(L1[k], L2[k])
       
        if (Cases[k] >= 0 and Cases[k] < 5000):
            earth.plot(x, y, 'ro',markersize=5,alpha=.5)
           
        elif (Cases[k] >= 5000 and Cases[k] < 140000):
            earth.plot(x, y, 'ro',markersize=((Cases[k])/1000),alpha=.5)
    k += 1
#This part outputs the plot that I designed above.
plt.title("Measles outbreaks (100559637)") #title
plt.show()

# Using the data we learned in class with the earthquake data I came up with a code.
#This part determines the colours and line width of the plot
#This plots the outbreak map for rubella
plt.figure(figsize=(10, 10))
earth = Basemap(projection='mill')

#this allows the different colours for the continents and the coastlines
earth.drawcoastlines(color='0.50', linewidth=0.25)
earth.fillcontinents(color='0.95')

# This opens the necessary file with the data needed.
# it puts the data into vectors that are needed to plot the data.
with open('Source_data_for_CFR_vaccine_map_abridged.csv') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        Cat.append(row['Category'])
        L1.append((float)(row['Latitude']))
        L2.append((float)(row['Longitude']))
        Cases.append((float)(row['Cases']))

#This part goes through all the values and the vectors and runs through the for loop and elif if statements.
# Then based on the values inputted into the loop the proper colours and dots will be outputted.
list(map(float, L1))
list(map(float, L2))
list(map(float, Cases))
l = 0
for i in range (l, len(Cat)):
    if (Cat[l] == 'Rubella'):
        x, y = earth(L1[l], L2[l])
       
        if(Cases[l] >= 0 and Cases[l] < 5000):
            earth.plot(x, y, 'mo',markersize=5,alpha=.5)
           
        elif(Cases[l] >= 5000 and Cases[l] < 140000):
            earth.plot(x, y, 'mo',markersize=((Cases[l])/1000),alpha=.5)
    l += 1     

#This part outputs the plot that I designed above.
plt.title("Rubella outbreaks (100559637)") #title
plt.show()


# Using the data we learned in class with the earthquake data I came up with a code.
#This part determines the colours and line width of the plot
#This plots the outbreak map for the mumps
plt.figure(figsize=(10, 10))
earth = Basemap(projection='mill')

#this allows the different colours for the continents and the coastlines
earth.drawcoastlines(color='0.50', linewidth=0.25)
earth.fillcontinents(color='0.95')

# This opens the necessary file with the data needed.
# it puts the data into vectors that are needed to plot the data.
with open('Source_data_for_CFR_vaccine_map_abridged.csv') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        Cat.append(row['Category'])
        L1.append((float)(row['Latitude']))
        L2.append((float)(row['Longitude']))
        Cases.append((float)(row['Cases']))

#This part goes through all the values and the vectors and runs through the for loop and elif if statements.
# Then based on the values inputted into the loop the proper colours and dots will be outputted.
list(map(float, L1))
list(map(float, L2))
list(map(float, Cases))
m = 0
for i in range (m, len(Cat)):
    if (Cat[m] == 'Mumps'):
        x, y = earth(L1[m], L2[m])
       
        if (Cases[m] >= 0 and Cases[m] < 5000):
            earth.plot(x, y, 'go',markersize=5,alpha=.5)
           
        elif(Cases[m] >= 5000 and Cases[m] < 140000):
            earth.plot(x, y, 'go',markersize=((Cases[m])/1000),alpha=.5)
    m += 1     

#This part outputs the plot that I designed above.
plt.title("Mumps outbreaks (100559637)") #title
plt.show()

# Using the data we learned in class with the earthquake data I came up with a code.
#This part determines the colours and line width of the plot
#This plots the outbreak map for polio
plt.figure(figsize=(10, 10))
earth = Basemap(projection='mill')

#this allows the different colours for the continents and the coastlines
earth.drawcoastlines(color='0.50', linewidth=0.25)
earth.fillcontinents(color='0.95')

# This opens the necessary file with the data needed.
# it puts the data into vectors that are needed to plot the data.
with open('Source_data_for_CFR_vaccine_map_abridged.csv') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        Cat.append(row['Category'])
        L1.append((float)(row['Latitude']))
        L2.append((float)(row['Longitude']))
        Cases.append((float)(row['Cases']))

#This part goes through all the values and the vectors and runs through the for loop and elif if statements.
# Then based on the values inputted into the loop the proper colours and dots will be outputted.
list(map(float, L1))
list(map(float, L2))
list(map(float, Cases))
n = 0
for i in range (n, len(Cat)):
    if (Cat[n] == 'Polio'):
        x, y = earth(L1[n], L2[n])
        if (Cases[n] >= 0 and Cases[n] < 5000):
            earth.plot(x, y, 'bo',markersize=5,alpha=.5)
           
        elif(Cases[n] >= 5000 and Cases[n] < 140000):
            earth.plot(x, y, 'bo',markersize=((Cases[n])/1000),alpha=.5)
    n += 1     

#This part outputs the plot that I designed above.   
plt.title("Polio outbreaks (100559637)") #title
plt.show()
```
