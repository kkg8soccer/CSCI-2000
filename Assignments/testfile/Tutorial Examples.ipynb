```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

n = 256
X = np.linspace(-np.pi,np.pi,n,endpoint=True)
Y = np.sin(2*X)

plt.plot (X, Y+1, color='blue', alpha=1.00)
plt.fill_between(X,1,Y+1, color = 'blue', alpha =0.5)
plt.plot (X, Y-1, color='red', alpha=1.00)
plt.fill_between(X,-1,Y-1,(Y-1)>-1, color = 'blue', alpha =0.5)
plt.fill_between(X,-1,Y-1,(Y-1)<-1, color = 'red', alpha =0.5)
plt.show()
```

```python
n = 12
X = np.arange(n)
Y1 = (1-X/float(n)) * np.random.uniform(0.5,1.0,n)
Y2 = (1-X/float(n)) * np.random.uniform(0.5,1.0,n) # spits out random values

plt.bar(X, +Y1, facecolor='#9999ff', edgecolor='white')
plt.bar(X, -Y2, facecolor='#ff9999', edgecolor='white')

for x,y in zip(X,Y1):
    plt.text(x+0.4, y+0.05, '%.2f' % y, ha='center', va= 'bottom') #%.2f gives two decimals
for x,y in zip(X,Y2):
    plt.text(x+0.4, -y-0.05, '%.2f' % -y, ha='center', va= 'top') # changes where the number are shown and if negative 
# plt.xticks([1,2, 2*np.pi)], ["1", "2", "$2|pi$"]) # erases the ticks of the plot
plt.xticks([])
plt.ylim(-1.25,+1.25)
plt.show()
```

```python
def f(x,y): return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2) #defining a function
#contour plot has levels, has each level as a colour, start with red, gets lighter as it fades with levels

n = 256
x = np.linspace(-3,3,n) # grid between -3 and 3 and an array of 256 points (for n)
y = np.linspace(-3,3,n)
X,Y = np.meshgrid(x,y) # x will be rows of x and y will be columns of y, is a matrix of 256x256

plt.contourf(X, Y, f(X,Y), 16 , alpha=.75, cmap='hot') # 16 is 16 different levels
C = plt.contour(X, Y, f(X,Y), 16, colors='black', linewidth=.5)
plt.clabel(C, fontsize = 12)
plt.colorbar()
plt.show()
```

```python
def f(x,y): return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)

n = 10
x = np.linspace(-3,3,4*n)
y = np.linspace(-3,3,3*n)
X,Y = np.meshgrid(x,y)
plt.imshow(f(X,Y), cmap = 'prism')
plt.xticks([])
plt.yticks([])
plt.show()

```
