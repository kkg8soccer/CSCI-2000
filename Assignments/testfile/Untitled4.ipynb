```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from JSAnimation.IPython_display import display_animation
```

```python
import urllib
from mpl_toolkits.basemap import Basemap
import csv
Category = []
Lat = []
Long = []
Cases = []

fig = plt.figure(figsize=(10,10))
ax = plt.subplot(1,1,1)

earth = Basemap(projection='mill')

earth.drawcoastlines(color='0.50', linewidth=0.25)
earth.fillcontinents(color='0.95')

with open('Source_data_for_CFR_vaccine_map_abridged.csv') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        Category.append(row['Category'])
        Lat.append((float)(row['Lat']))
        Long.append((float)(row['Long']))
        Cases.append((float)(row['Cases']))
        
list(map(float, Lat))
list(map(float, Long))
list(map(float, Cases))
k = 0
for i in range (k, len(Category)):
    if (Category[k] == 'Measles'):
        x, y = earth(Long[k], Lat[k])
        
        if (Cases[k] >= 0 and Cases[k] < 5000):
            earth.plot(x, y, 'ro',markersize=5,alpha=.5)
            
        elif (Cases[k] >= 5000 and Cases[k] < 140000):
            earth.plot(x, y, 'ro',markersize=((Cases[k])/1000),alpha=.5)
    k += 1
    
plt.title("Measles outbreaks (100555440)")
plt.show()
```
