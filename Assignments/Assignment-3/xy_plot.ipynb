```python
# These are the necessary libraries for this question
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# This imports the file that has the data we are working with
file=open('xy_points.dat','r')

line=file.readline() #file.readline() reads an entire line from the file
x=line.split() #line.split() returns a list of all the words in the string
x=np.array(x) #np.array(y) creates an array called x with all the x values in it


line=file.readline()
y=line.split()
y=np.array(y) #np.array(y) creates an array called y with all the y values in it

x1=[]
x2=[]
y1=[]
y2=[] #Here I am defining my variables

plt.plot(x,y,'ro') #This plots all the points (x,y) from the file xy_points.dat as red dots

#This part shows the blue region that is in x<-2 and y>=-5 and y<=0

for i in range(len(x)): # This part goes through the whole length of the array x
    if float(x[i])<-2 and float(y[i])>=-5.0 and float(y[i])<=0.0:
        x1.append(x[i])
        y1.append(y[i]) # This adds the values to be changed colours
        
plt.plot(x1,y1,'bo') # This plots all the dots that are in the above domain as blue dots


#This part shows the green region that is bound by the circle region of radius 5 of the point (0,5)

for i in range(len(x)):
    if (float(x[i])**2+(float(y[i])-5)**2)<25: # the ** is used for raised to power, this is the equation of a circle that is the bound of the plot points. 
    # it must be < 25 because it is the range in the circle   
        x2.append(x[i])
        y2.append(y[i])  # This adds the values to be changed colours
        
plt.plot(x2,y2,'go') # This plots all the dots that are in the above domain as green dots

plt.title('Highlighted points (100559637)') #This part plots the title of the graph
plt.axis([-10, 10, -10, 10]) # This part makes the square region of the axes [-10,10]x[-10,10]

plt.show() # This plots what we have created above
```
